```python
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, Input
from tensorflow.keras.optimizers import RMSprop,Adam
import tensorflow as tf
```

```python
from tensorflow.keras.layers import BatchNormalization
#IMG_SIZE = 32
input = Input(shape=(768,))

# x = LSTM(SEQ_LEN)(lstm_input)


xF = Dense(512, activation = "relu")(input)
xF = BatchNormalization()(xF)

xF = Dense(512, activation = "relu")(xF)
xF = BatchNormalization()(xF)
xF = Dropout(0.5)(xF)

xF = Dense(512, activation = "relu")(xF)
xF = BatchNormalization()(xF)
xF = Dropout(0.5)(xF)

xF = Dense(256, activation = "relu")(xF)
xF = BatchNormalization()(xF)
xF = Dropout(0.5)(xF)

xF = Dense(64, activation = "relu")(xF)
xF = BatchNormalization()(xF)
xF = Dropout(0.5)(xF)

xF = Dense(16, activation = "relu")(xF)
xF = BatchNormalization()(xF)
xF = Dropout(0.5)(xF)

output = Dense(4, activation = "softmax")(xF)
print(output)
model = Model(inputs=input, outputs=output)


optimizer = Adam(learning_rate=0.01, decay=1e-6) #RMSprop(learning_rate=0.001, rho=0.9, epsilon=1e-08, decay=0.0)#
model.compile(optimizer = optimizer , loss = "binary_crossentropy",loss_weights=[1.0], metrics=["accuracy"])

# model.summary()
# tf.keras.utils.plot_model(model,  show_shapes=True)

# x = CreateLSTMPart(lstm_input)
#   #x = Dense(128, name='dense_2', activation='relu')(Flatten()(model_input))
#   x = Dropout(0.2)(x)
#   lstm_branch = Model(inputs=lstm_input, outputs=x)

#   indy_input = Input(shape=(TechFeatureCnt + 4))#FeatureCnt - LSTMFeatureCnt))
#   y = Dense(20)(indy_input)
#   y = Activation("relu")(y)
#   y = Dropout(0.2)(y)
#   technical_indicators_branch = Model(inputs=indy_input, outputs=y)

#   combined = concatenate([lstm_branch.output, technical_indicators_branch.output])

#   #x = Dense(128, name='dense_3')(x)
#   #x = Activation('sigmoid', name='sigmoid_1')(x)
#   z = Dense(64, activation="sigmoid")(combined)
#   z = Dense(1, activation="linear")(z)

#   return Model(inputs=[lstm_branch.input, technical_indicators_branch.input], outputs=z)

model.summary()
tf.keras.utils.plot_model(model,  show_shapes=True)
```

```python
from tqdm import tqdm
import tensorflow as tf
class ProgressBar(tf.keras.callbacks.Callback):
    def __init__(self,EPOCH):
        self.ProgressBar = tqdm(total=EPOCH,ascii=True)
        self.minloss=0
#         self.minloss=1e10
        self.minpos=0
        self.min_n = 0
        pass
    def on_epoch_end(self, e, logs=None):
#         if(logs['val_loss']<self.minloss):
#           self.minloss = logs['val_loss']
#           self.min_n = self.ProgressBar.n
#         self.ProgressBar.set_description(f"Loss = {logs['loss']:.3e} Val_loss = {logs['val_loss']:.3e} min = {self.ProgressBar.n-self.min_n}:{self.minloss:.3e}")
        if(logs['val_accuracy']>self.minloss):
          self.minloss = logs['val_accuracy']
          self.min_n = self.ProgressBar.n  
        self.ProgressBar.set_description(f"accuracy = {logs['accuracy']:.3e} Val_accuracy = {logs['val_accuracy']:.3e} max = {self.ProgressBar.n-self.min_n}:{self.minloss:.3e}")
        self.ProgressBar.update(1)
        pass
    def on_train_end(self,logs=None):
        self.ProgressBar.close()
        pass
```
